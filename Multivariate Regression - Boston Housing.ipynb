```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
housing_data = load_boston()
```

```python
print(housing_data.keys())
```

```python
print(housing_data['feature_names'])
```

```python
print(housing_data['DESCR'])
```

```python
boston = pd.DataFrame(housing_data.data) #using pandas printed housig data in a format
boston.columns = housing_data.feature_names # assigning the features names to data frame
boston.head(9)
```

```python
#Adding price which is the target to the data frame
boston['Price'] = housing_data.target
boston.head()
```

```python
boston.describe()
```

```python
sns.heatmap(boston.corr()) #relation between different features
```

```python
# creating two arrays one dependent and one independent
x = boston.drop('Price',axis=1)
y = boston['Price']
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3) 
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
theta = np.zeros(x_train.shape[1])
print(theta.shape)
print(len(y_train))
```

```python
#trainig the train part of the module using gradient descent
theta = np.zeros(x_train.shape[1])
bias = 0
l=0.2
epochs = 500

n=x_train.shape[0]
m=x_test.shape[0]
costs_train=[]
m_train = y_train.shape[0]
m_test = y_test.shape[0]

for i in range(epochs):
    y_pred = np.dot(x_train,theta)+bias
    cost_train = (1/(2*n))*np.sum(np.square(y_pred-y_train))
    dz = (1/n)*(y_pred-y_train)
    dw = np.dot(dz.T,x_train)
    db = np.sum(dz)
    theta = theta - l*dw
    bias = bias - l*db
    
    if(i%10==0):
        costs_train.append(cost_train)
    
    #MAE (Mean absolute error) train
    MAE= (1/m_train)*np.sum(np.abs(y_pred-y_train))
    
    #prediction
    y_pred_test = np.dot(x_test,theta)+bias
    cost_test = (1/(2*m))*np.sum(np.square(y_pred_test-y_test)) 
    MAE_test = (1/m_test)*np.sum(np.abs(y_pred_test-y_test))
    
    #printing values
    print("Epochs"+str(i))
    print("Training cost :"+str(cost_train)+ "| Validtaion Cost :"+ str(cost_test))
    print("Training MAE :"+str(MAE)+ "| MAE test :"+ str(MAE_test))

plt.plot(costs_train)
plt.show()
```

```python
#comparing our model with sklearn model
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
# print the intercept
print(lr.intercept_)
```

```python
pred = lr.predict(x_test)
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, pred))
```
